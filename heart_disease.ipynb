```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df = pd.read_csv('./Heart_Disease_Prediction.csv')
df
```

```python

X = df.drop(columns=['Heart Disease'], axis=1)
Y = df['Heart Disease']
Y
```

```python
plot_number = 1
plt.figure(figsize=(15,25))
columns = df.columns
for i in columns:
     if i not in ['Age', 'ST depression', 'Max HR', 'Cholesterol', 'BP']:
        plt.subplot(5, 2, plot_number)
        sns.countplot(x=i, data=df)
        plot_number = plot_number + 1

plt.show()
```

```python
plot_number = 1
plt.figure(figsize=(15,25))
columns = df.columns
for i in columns:
        plt.subplot(7, 2, plot_number)
        sns.histplot(x=i, data=df)
        plot_number = plot_number + 1

plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.shape[1]
y_test
```

```python
# model = tf.keras.Sequential([
#     tf.keras.layers.Dense(1, input_dim=X_train.shape[1], activation='sigmoid')
# ])
model = tf.keras.Sequential([
    tf.keras.layers.Dense(14, input_dim=X_train.shape[1], activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python

model.fit(X_train, y_train, epochs=256, batch_size=32, validation_data=(X_test, y_test))
```
